# Problem 1: Unique Words in a Sentence

```python
# Given Sentence
sentence = "python is easy and python is powerful"

# Split the sentence into a list of words
words = sentence.split()

# Convert the list to a set to remove duplicates
unique_words = set(words)

# Count how many items are in the set
count = len(unique_words)

# Display the results
print(f"Unique words count: {count}")
print(f"Unique words: {unique_words}")
```

# Problem 2: Highest Salary from Employee Data


```python
# Given Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Initialize variables to track the best so far
max_salary = 0
top_employee = ""

# Loop through the dictionary items
for name, salary in employees.items():
    if salary > max_salary:
        max_salary = salary
        top_employee = name

# Print the Highest Salary Employee
print(f"Highest Salary: {top_employee} - {max_salary}")
```

# Problem 3: Find Maximum and Minimum Values


```python
# Given Numbers
numbers = [45, 22, 89, 10, 66]

# Start by assuming the first number is both the max and min
max_val = numbers[0]
min_val = numbers[0]

# Loop through the list to compare
for n in numbers:
    # Update max if we find a bigger number
    if n > max_val:
        max_val = n

    # Update min if we find a smaller number
    if n < min_val:
        min_val = n

# Display results
print(f"List: {numbers}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
```

# Problem 4: Count Products Above a Price Threshold


```python
# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Start our counter at 0
count = 0

# Look at each price in the list
for p in prices:
    # Check if the price is above 1000
    if p > 1000:
        count += 1 # Add 1 to our counter if true

# Show the final result
print(f"Products above 1000: {count}")
```

#Problem 5: Calculate Attendance Percentage


```python
# Given Attendence

attendance = ["P", "P", "A", "P", "P"]

# Count the number of "P" (Present) entries
present_count = attendance.count("P")

# Get the total number of days in the list
total_days = len(attendance)

# Calculate the percentage
# (Presents divided by Total) multiplied by 100
percentage = (present_count / total_days) * 100

# Show the final result
print(f"Attendance Percentage: {percentage}")
```

#Problem 7: Count Character Frequency


```python
# Given Data
word = "pythonp"

# Create an empty dictionary to store counts
frequency = {}

# Loop through each character in the string
for char in word:
    if char in frequency:
        # If the character is already in the dictionary, add 1
        frequency[char] += 1
    else:
        # character is set it to 1
        frequency[char] = 1

# Display the result
print(frequency)
```

# Problem 9: Check if a Key Exists in a Dictionary

```python
# Our dictionary of employee data
employees = {"ID101": "Ravi", "ID102": "Anita", "ID103": "Kiran"}

The key we want to look for
search_key = "ID102"

# Use the 'in' keyword to check
if search_key in employees:
    print("Employee exists")
else:
    print("Employee not found")
```

#Problem 10: Calculate Average Marks


```python
# Given Data
marks = [80, 65, 90, 55]

# 1. Calculate the sum of all marks
total_sum = sum(marks)

# 2. Find out how many subjects there are
total_subjects = len(marks)

# 3. Divide sum by count to get the average
average = total_sum / total_subjects

# 4. Display the result
print(f"Average Marks: {average}")
```

#1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement():
    # User need to enter the data
    print("Social Media Post Analyzer")
    data_input = input("Enter likes received (separated by spaces): ")

    # Converting the input string into a list of integers
    likes_list = list(map(int, data_input.split()))

    # 0 is the initial value for calculation
    total = 0
    for like in likes_list:
      # Adding each value to the total
        total += like

    # Engagement Logic based on the 'total' variable
    if total >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Print Expected Output Format

    print(f"Total Likes: {total}")
    print(f"Post Status: {status}")

# Call the function to start the process
analyze_post_engagement()
```

#3: Agriculture – Rainfall Adequacy Checker


```python
# Function to analyze rainfall data
def check_rainfall():
    n = int(input("Enter number of rainfall records: "))

    rainfall_data = []

    for i in range(n):
        rain = float(input(f"Enter rainfall value {i+1}: "))
        rainfall_data.append(rain)

    required_level = float(input("Enter required rainfall level: "))

    average = sum(rainfall_data) / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("\nAverage Rainfall:", average)
    print("Rainfall Status:", status)

# Call the function to start the process
check_rainfall()
```

#4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts():
    print("Social Media: Duplicate Account Detection")

    # User enters usernames (e.g., user1 user2 user1 user3)
    data_input = input("Enter usernames (separated by spaces): ")

    # Convert input into a list
    usernames_list = data_input.split()

    # Use set data structure
    # Sets automatically remove duplicates
    unique_usernames = set(usernames_list)

    if len(usernames_list) != len(unique_usernames):
        duplicate_found = "Yes"
    else:
        duplicate_found = "No"

    # Expected Output Format

    print(f"Duplicate Accounts Found: {duplicate_found}")

# Call the function to start the process
detect_duplicate_accounts()
```

#6: Agriculture – Premium Crop Price Filter


```python
def premium_crop_filter():
    # 1. User enters data (e.g., 1500, 2500, 3200)
    raw_data = input("Enter crop prices separated by commas: ")

    # 2. Split the string by comma and convert each part to an integer
    # This creates a list: [1500, 2500, 3200]
    try:
        prices = [int(p.strip()) for p in raw_data.split(',')]

        # 3. Filter the premium crops (> 2000)
        premium_crops = [price for price in prices if price > 2000]

        # Expected Output Format
        print(f"Premium Crops: {premium_crops}")

    except ValueError:
        print("Error: Please ensure you only enter numbers separated by commas.")

# Call the function to start the process
premium_crop_filter()
```

#7: System Monitoring – Application Health Checker


```python
def check_system_health():
    try:
        # User enters the number of errors detected
        error_count = int(input("Enter the number of system errors: "))

        # Logic Flow: Starting from the most specific to the most general
        if error_count == 0:
            status = "Healthy"
        elif error_count <= 5:
            status = "Minor Issues"
        else:
            status = "Critical Issues"

        # Expected Output Format
        print(f"Error Count: {error_count}")
        print(f"System Status: {status}")

    except ValueError:
        print("Error: Please enter a valid whole number for the error count.")

# Call the function to start the process
check_system_health()
```

#8: Banking – Daily Transaction Limit Checker


```python
def check_transaction_limit():
    # Constant daily limit
    DAILY_LIMIT = 50000

    try:
        # User enters the transaction amount
        amount = float(input("Enter the transaction amount: "))

        # Validation Logic
        if amount <= DAILY_LIMIT:
            status = "Approved"
        else:
            status = "Rejected"

        # Expected Output Format
        print(f"Transaction Amount: {amount}")
        print(f"Transaction Status: {status}")

    except ValueError:
        print("Error: Please enter a valid numerical amount.")

# Call the function to start the process
check_transaction_limit()
```

#9: E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility():
    try:
        # User enters attendance as 1s and 0s (e.g., 1, 0, 1, 1)
        raw_data = input("Enter attendance (1 for Present, 0 for Absent) separated by commas: ")

        # Convert input into a list of integers
        attendance_list = [int(x.strip()) for x in raw_data.split(',')]

        # Calculate Percentage: (Sum of presence / Total days) * 100
        total_days = len(attendance_list)
        total_present = sum(attendance_list)

        if total_days == 0:
            print("No attendance records found.")
            return

        percentage = (total_present / total_days) * 100

        # Eligibility Check
        if percentage >= 75:
            status = "Eligible"
        else:
            status = "Not Eligible"

        # Expected Output Format
        print(f"\nAttendance Percentage: {percentage:.1f}")
        print(f"Exam Eligibility: {status}")

    except ValueError:
        print("Error: Please enter only 1 or 0 separated by commas.")

# Call the function to start the process
check_exam_eligibility()
```

#10: Smart Electricity Bill Analyzer


```python
def electricity_bill_analyzer():
    try:
        units = int(input("Enter units consumed: "))
        bill = 0

        # Step 1: Calculate Bill using Slab Logic
        if units <= 100:
            bill = units * 3
        elif units <= 200:
            # First 100 units at ₹3 + remaining units at ₹5
            bill = (100 * 3) + ((units - 100) * 5)
        else:
            # First 100 at ₹3 + next 100 at ₹5 + remaining at ₹7
            bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

        # Step 2: Classify Usage based on total Bill
        if bill < 500:
            status = "Low Usage"
        elif 500 <= bill <= 1500:
            status = "Moderate Usage"
        else:
            status = "High Usage"

        # Output the results
        print(f"\nUnits Consumed: {units}")
        print(f"Total Bill: ₹{bill}")
        print(f"Usage Status: {status}")

    except ValueError:
        print("Error: Please enter a valid number of units.")

# Execute the function
electricity_bill_analyzer()
```

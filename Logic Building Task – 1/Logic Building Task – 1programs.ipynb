#2. Pass or Fail Analyzer

```python
# This is our list of student scores
marks = [45, 78, 90, 33, 60]

# We start our count at zero
pass_count = 0
fail_count = 0

# Look at each score one by one(Where Logic Works)
for m in marks:
    # Check if the score is 50 or higher
    if m >= 50:
        pass_count += 1  # Add 1 to pass total
    else:
        fail_count += 1  # Add 1 to fail total

# Show the final totals on the screen
print(f"Total Students Passed: {pass_count}")
print(f"Total Students Failed: {fail_count}")
```

#3. Simple Data Cleaner


```python
# List of Dummy names
names = [" nandish ", "ram", " dimpu "]

# Create an empty list to store the clean names
cleaned_names = []

# Loop starts with each name in the list
for n in names:
    # .strip() removes spaces from ends
    # .lower() makes all letters lowercase
    clean_n = n.strip().lower()

    # Add the clean names to our new list
    cleaned_names.append(clean_n)

# Print the final result
print("Cleaned List:", cleaned_names)
```

#4. Message Length Analyzer


```python
# Given Message
messages = ["Hi", "Welcome to the platform", "OK"]

# Process each message in the list
for msg in messages:
    # Get the number of characters
    length = len(msg)

    # Check if the message is longer than 10
    if length > 10:
        status = "Flagged: Too Long"
    else:
        status = "Short Message"

    # Print the results clearly
    print(f"Message: '{msg}' | Length: {length} | {status}")
```

#5. Error Message Detector

```python
# Given Logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Start counting from 0
error_count = 0

# Check each entry in the logs
for entry in logs:
    # If the entry matches exactly "ERROR"
    if entry == "ERROR":
        # Add 1 to our counter
        error_count += 1

# Print the total
print(f"Total Errors Found: {error_count}")
```

#1:Smart Parking Lot Management System

```python
def manage_parking(capacity, logs):
    current_vehicles = 0

    for action in logs:
        action = action.strip().upper()  # Clean the input

        if action == "IN":
            current_vehicles += 1
        elif action == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1

        # Alert if capacity exceeded during the process
        if current_vehicles > capacity:
            print(f"--- ALERT: Capacity exceeded! ({current_vehicles}/{capacity}) ---")

    # Determine final status
    status = "Full" if current_vehicles >= capacity else "Available"

    # Return results for printing
    return current_vehicles, status

# USER INPUT SECTION
print("Smart Parking System User Portal")

# Get Capacity
cap_input = int(input("Enter Parking Capacity: "))

# Get Logs (User enters words separated by commas)
log_input = input("Enter Vehicle Logs (e.g., IN, IN, OUT, IN): ")
# Convert the string "IN, IN, OUT" into a list ["IN", "IN", "OUT"]
user_logs = log_input.split(",")

# PROCESSING
parked, status = manage_parking(cap_input, user_logs)

# FINAL OUTPUT
print("\nFinal Report")
print(f"Currently Parked Vehicles: {parked}")
print(f"Parking Status: {status}")
```

#2:Online Food Delivery Time Estimator


```python
def estimate_delivery(distance, weather, traffic):

    base_time = (distance * 5)

    # WEATHER
    if weather == "Rainy":
        base_time += 5
    elif weather == "Snowy":
        base_time += 10

    # TRAFFIC
    if traffic == "High":
        base_time += 10
    elif traffic == "Medium":
        base_time += 5

    return int(base_time)  # Return must be at the very end of the function


# USER INPUT
print("Online Food Delivery System ")

# Get Distance (km)
dist_input = float(input("Enter Distance (km): "))

# Get Weather Condition
weather_input = input("Enter Weather Condition (Clear, Rainy, Snowy): ").capitalize()

# Get Traffic Level
traffic_input = input("Enter Traffic Level (Low, Medium, High): ").capitalize()


# PROCESSING
final_eta = estimate_delivery(dist_input, weather_input, traffic_input)

# FINAL OUTPUT
print("\nDelivery Estimate:")
print(f"Estimated Delivery Time: {final_eta} minutes")
```

#3:Movie Theatre Seat Occupancy Analyzer



```python
def analyze_theatre(total_seats, booked_count):
    # Calculate percentage
    occupied_percentage = (booked_count / total_seats) * 100

    # Determine status
    if occupied_percentage == 100:
        status = "Housefull"
    elif occupied_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    return occupied_percentage, status

# USER INPUT
print("Cinema Seat Occupancy Analyzer")

# Get inputs
total = int(input("Enter Total Seats: "))
booked = int(input("Enter Number of Booked Seats: "))

# Process
percent, status = analyze_theatre(total, booked)

# Output
print("\nResults")
print(f"Occupied: {int(percent)}%")
print(f"Show Status: {status}")
```

#4:Cloud Server Load Classification System


```python
def classify_server_load(readings):
    # Calculate the average CPU load
    average_load = sum(readings) / len(readings)

    # Classify the status based on the rules
    if average_load > 80:
        status = "Critical"
    elif 50 <= average_load <= 80:
        status = "Warning"
    else:
        status = "Normal"

    return average_load, status

# USER INPUT
print("Cloud Server Monitoring System")

# Get readings from user as a comma-separated string
raw_input = input("Enter CPU Readings (%) : ")

# Convert the string input into a list of numbers
cpu_readings = [int(i.strip()) for i in raw_input.split(",")]

# Process
avg, server_status = classify_server_load(cpu_readings)

# Output
print("\nMonitoring Report")
print(f"Average CPU Load: {int(avg)}%")
print(f"Server Status: {server_status}")
```

#5:Smart Classroom Resource Usage Monitor

```python
def check_usage(projector, ac, lights):
    # Set Limit To 8 Hours
    limit = 8
    overused = []

    # Check each resource individually
    if projector > limit:
        overused.append("Projector")
    if ac > limit:
        overused.append("AC")
    if lights > limit:
        overused.append("Lights")

    # Determine alert status
    alert = "Yes" if len(overused) > 0 else "No"

    return overused, alert

# USER INPUT
print("Smart Classroom Resource Monitor")

# Simple inputs
p_hours = int(input("Enter Projector hours: "))
a_hours = int(input("Enter AC hours: "))
l_hours = int(input("Enter Lights hours: "))

# Process
overused_list, energy_alert = check_usage(p_hours, a_hours, l_hours)

# Output
print("\n--- Summary ---")
# If the list is empty, show "None"
print(f"Overused Resources: {', '.join(overused_list) if overused_list else 'None'}")
print(f"Energy Alert: {energy_alert}")
```

#6: Online Event Registration Capacity Controller



```python
def manage_event_registration(capacity, total_registrations):
    # Initialize variables
    confirmed = 0
    waitlisted = 0

    # Logic to split registrations
    if total_registrations <= capacity:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = total_registrations - capacity
        status = "Closed"

    return confirmed, waitlisted, status

# USER INPUT
print("Online Event Registration System")

# Get Event Capacity
event_cap = int(input("Enter Event Capacity: "))

# Get Total Registrations
reg_count = int(input("Enter Total Registrations: "))

# PROCESSING
confirmed_users, waitlisted_users, reg_status = manage_event_registration(event_cap, reg_count)

# FINAL OUTPUT
print("\nRegistration Summary ")
print(f"Confirmed Registrations: {confirmed_users}")
print(f"Waitlisted Users: {waitlisted_users}")
print(f"Registration Status: {reg_status}")
```

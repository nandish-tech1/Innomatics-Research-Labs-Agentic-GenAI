#1: Mobile Recharge Validation System


```python
# Function for validation
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return "Invalid: Amount must be at least ₹50."
    elif amount not in valid_plans:
        return f"Invalid: {amount} is not in the valid plans: {valid_plans}"
    else:
        return "Valid"

# While loop for retry
while True:
    user_amount = int(input("Enter recharge amount: ₹"))

    result = validate_recharge(user_amount)

    if result == "Valid":
        print(f"Success! Recharge of ₹{user_amount} completed.")
        break  # Exit the loop on success
    else:
        print(result)
        print("Please try again.\n")
```

#2: Inventory Reorder Alert System


```python
# Given Inventory Data and maintain Stock Quantity
inventory = {
    "Laptops": 26,
    "Mice": 11,
    "Keyboards": 15,
    "Monitors": 29,
    "USB Cables": 7
}

# Function to check stock level
def check_stock(product, quantity):
    if quantity < 15:
        print(f"REORDER ALERT: {product} is low (Stock: {quantity})")
    else:
        print(f"Stock OK: {product} (Stock: {quantity})")

# Use for loop on dictionary
print("Inventory Status Report:")
for product, quantity in inventory.items():
    check_stock(product, quantity)
```

#3: Student Result Processing

```python
# Function to process student results
def process_result(marks_list):
    total = 0

    # Hint: Use loop to calculate total
    for mark in marks_list:
        total = total + mark

    average = total / len(marks_list)

    # Return result from function
    if average >= 50:
        return "Pass", average
    else:
        return "Fail", average

# Logic to take user input
user_marks = []
print("Enter marks one by one. Type 'done' when finished.")

while True:
    entry = input("Enter mark: ")

    if entry.lower() == 'done':
        break

    # Convert input string to integer and add to list
    try:
        mark = int(entry)
        user_marks.append(mark)
    except ValueError:
        print("Invalid input. Please enter a number or 'done'.")

# Check if list is not empty before processing
if len(user_marks) > 0:
    status, avg = process_result(user_marks)
    print(f"\nAverage: {avg}")
    print(f"Result: {status}")
else:
    print("No marks were entered.")
```

#4: Cab Fare Estimator with Retry Option


```python
# Function for fare calculation
def calculate_fare(distance, is_peak):
    base_fare = 50
    rate_per_km = 12

    # Calculate initial fare
    total_fare = base_fare + (distance * rate_per_km)

    # Apply 25% extra if it is a peak hour
    if is_peak.lower() == 'yes':
        total_fare = total_fare * 1.25

    return total_fare

# While loop for retry
while True:
    print("\nCab Fare Estimator")
    dist = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    fare = calculate_fare(dist, peak)
    print(f"Estimated Total Fare: ₹{fare:.2f}")

    # Check if user wants to calculate another fare
    retry = input("\nDo you want to calculate again? (yes/no): ")
    if retry.lower() != 'yes':
        print("Thank you for using the Cab Fare Estimator!")
        break
```

#5: Employee Attendance Eligibility Checker

```python
# Function to check eligibility
def check_eligibility(attendance_list):
    present_days = 0
    total_days = len(attendance_list)

    # Hint: Use loop to count present days
    for status in attendance_list:
        if status.upper() == "P":
            present_days += 1

    # Calculate percentage
    attendance_percent = (present_days / total_days) * 100

    # Logic for eligibility (Threshold 75%)
    if attendance_percent >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"

    # Hint: Use return statement
    return result, attendance_percent

# Input: Attendance list (P = Present, A = Absent)
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

# Calling the function
status, percentage = check_eligibility(attendance)

# Output
print(f"Attendance Percentage: {percentage}%")
print(f"Status: {status}")
```

#6: Password Strength Checker

```python
def check_password_strength(password):
    # Minimum length check
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long."

    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Use loop over string
    for char in password:
        # Check for digits
        if char.isdigit():
            has_digit = True

        # Check for special characters
        if char in special_chars:
            has_special = True

    # Use conditions inside function
    if not has_digit:
        return "Weak: Must contain at least one digit (0-9)."

    if not has_special:
        return "Weak: Must contain at least one special character (@, #, or $)."

    return "Strong Password!"

# Taking user input
user_pass = input("Create a password: ")
result = check_password_strength(user_pass)
print(result)
```

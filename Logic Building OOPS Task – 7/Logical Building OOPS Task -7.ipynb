#1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name, status):
        self.name = name
        self.status = status

    def control(self, action):
        if action == "ON":
            self.status = "ON"
            print(f"{self.name} is ON")
        else:
            self.status = "OFF"
            print(f"{self.name} is OFF")

    def display_status(self):
        print(f"ACTION: {self.status}")


## Create light object
light = SmartLight("Bedroom Light", "ON")

## Control ON/OFF
action = "ON"

light.control(action)
light.display_status()
```

#2: Employee ID Card System

```python
class EmployeeIDCard:

  # Create employee
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    # Display ID card
    def display_idcard(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.dept}")

# input values
emp = EmployeeIDCard("Nandish", "EMP102", "IT")

emp.display_idcard()
```

#3: Mobile Contact Record


```python
class MobileContact:
    def __init__(self, name, phone_number):  # Fixed def, comma, snake_case
        self.name = name
        self.phone_number = phone_number  # Fixed variable name

    def display_contact(self):  # Proper indentation
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Mobile Number: {self.phone_number}")

# Fixed class name and method call
contact = MobileContact("Nandish", "4526171818")
contact.display_contact()
```

#4: Product Price Tag Generator

```python
class ProductPriceTag:

  # Create product
    def __init__(self, name, price):
        self.name = name
        self.price = price

      # Display price tag
    def print_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

# input values
product = ProductPriceTag("OnePlus Node ce 5", 26799)

product.print_tag()
```

#5: Movie Rating Display System


```python
class MovieRating:

  # Create movie
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

     # display movie details
    def display_movie(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

# input values
movie = MovieRating("Peddi ", 4.9)

movie.display_movie()
```

#6: Delivery Address Manager


```python
class DeliveryAddress:

   # Create delivery
    def __init__(self, name, address):
        self.name = name
        self.address = address

    # Show delivery info
    def print_details(self):
        print("Delivery Details")
        print(f"Customer: {self.name}")
        print(f"Address: {self.address}")

#  input values
delivery = DeliveryAddress("Nandish", "Bengaluru")

delivery.print_details()
```

# 1: Employee Performance Bonus Eligibility


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# 1. Find the highest score value in the dictionary
highest_score = max(employees.values())

# 2. Create a list of all names that match the highest score
top_performers = []
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# 3. Join the names into a string for clean display
names_str = ", ".join(top_performers)

# 4. Final Output
print(f"Top Performers Eligible for Bonus: {names_str} (Score: {highest_score})")#
```

#2: Search Query Keyword Analysis

```python
# The input string
query = "Buy mobile phone buy phone online"

# Preprocess: Make lowercase and split into words
words = query.lower().split()

# Create a dictionary to store counts
counts = {}
for w in words:
    counts[w] = counts.get(w, 0) + 1

# Filter for words that appear more than once
# This creates the final dictionary using a simple loop
result = {}
for word, count in counts.items():
    if count > 1:
        result[word] = count

# Output the result
print(result)
```

#3: Sensor Data Validation


```python
# Input list of sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Using a list comprehension to filter even readings
# 'i' represents the hour index, 'val' represents the reading
valid_readings = [(i, val) for i, val in enumerate(sensor_readings) if val % 2 == 0]

# Displaying the results
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# 4: Email Domain Usage Analysis


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_counts = {}
total_emails = len(emails)

# Process each email
for email in emails:
    # Split the email at '@' and take the second part
    domain = email.split('@')[1]

    # Update the count in our dictionary
    domain_counts[domain] = domain_counts.get(domain, 0) + 1

# Calculate and display percentages
print("Email Domain Usage Analysis:")
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

#5: Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate the average daily sales
average_sales = sum(sales) / len(sales)

# Define the spike threshold (30% above average)
threshold = average_sales * 1.30

print(f"Average Sales: {average_sales:.2f}")
print(f"Spike Threshold: {threshold:.2f}")
print("-" * 25)

# Detect and display spikes
# We use i+1 because list indices start at 0, but days start at 1
for i, value in enumerate(sales):
    if value > threshold:
        print(f"Day {i + 1}: {value}")
```

#6: Duplicate User ID Detection


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Create a dictionary to store the frequency of each user ID
id_counts = {}

for uid in user_ids:
    # Increment the count if the ID exists, otherwise start at 1
    id_counts[uid] = id_counts.get(uid, 0) + 1

# Identify and display only the duplicates (count > 1)
print("Duplicate User ID Detection:")
for uid, count in id_counts.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
